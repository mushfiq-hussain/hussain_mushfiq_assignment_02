# Question 1 (40 points)

Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(2) Collect the top 1000 User Reviews of a movie recently in 2023 or 2024 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]

(3) Collect all the reviews of the top 1000 most popular software from G2 or Capterra.

(4) Collect the **abstracts** of the top 10000 research papers by using the query "machine learning", "data science", "artifical intelligence", or "information extraction" from Semantic Scholar.

(5) Collect all the information of the 904 narrators in the Densho Digital Repository.


```python
import requests
from bs4 import BeautifulSoup
import csv

# Product URL
url = "https://www.amazon.com/gp/aw/d/B0CHBPNBSM/?_encoding=UTF8&pd_rd_plhdr=t&aaxitk=58823e8b2de10749f66f2c476aec2054&hsa_cr_id=0&qid=1708907658&sr=1-1-9e67e56a-6f64-441f-a281-df67fc737124&ref_=sbx_be_s_sparkle_lsi4d_asin_0_img&pd_rd_w=O7grp&content-id=amzn1.sym.417820b0-80f2-4084-adb3-fb612550f30b%3Aamzn1.sym.417820b0-80f2-4084-adb3-fb612550f30b&pf_rd_p=417820b0-80f2-4084-adb3-fb612550f30b&pf_rd_r=S5NPKN6VFD83CKY60Y52&pd_rd_wg=ZHArD&pd_rd_r=5f35555e-65f0-44ae-a39e-96fac19563be"

# List to store reviews
reviews = []

# Iterate through pagination
for i in range(1, 11):

  # Request page
  headers = {"User-Agent": "Mozilla/5.0"}
  r = requests.get(url + "&pageNumber=" + str(i), headers=headers)

  # Parse HTML
  soup = BeautifulSoup(r.content, 'html.parser')

  # Get reviews on page
  product_reviews = soup.find_all('div', {'data-hook': 'review'})

  # Extract review text
  for review in product_reviews:
    reviews.append(review.find('span', {'data-hook':'review-body'}).text.strip())

  print(f"Page {i} done. Reviews extracted: {len(reviews)}")

# Write reviews to CSV file
with open('iphone_reviews.csv', 'w', newline='') as f:
  writer = csv.writer(f)
  writer.writerow(['review'])
  for review in reviews:
    writer.writerow([review])

print(f"Total reviews extracted: {len(reviews)}")
```

# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
import csv
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

with open('iphone_reviews.csv', 'r') as infile, open('cleaned_reviews.csv', 'w') as outfile:

  reader = csv.reader(infile)
  writer = csv.writer(outfile)

  for row in reader:

    text = row[0]

    # (1) Remove noise
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    print("(1)", text)

    # (2) Remove numbers
    text = re.sub(r'\d+', '', text)
    print("(2)", text)

    # (3) Remove stopwords
    text = ' '.join([word for word in text.split() if word not in stop_words])
    print("(3)", text)

    # (4) Lowercase
    text = text.lower()
    print("(4)", text)

    # (5) Stemming
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    print("(5)", text)

    # (6) Lemmatization
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
    print("(6)", text)

    row.append(text)
    writer.writerow(row)

print("Cleaning done!")
```
